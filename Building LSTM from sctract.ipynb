```python
import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Tanh activation function
def tanh(x):
    return np.tanh(x)

# Input data
x_t = np.array([0.5])  # Current input
h_prev = np.array([0.1])  # Previous hidden state
C_prev = np.array([0.2])  # Previous cell state

# LSTM Weights and Biases (updated to match the input dimensions)
W_f = np.array([0.8, 0.8])  # Forget gate weight (updated shape to (2,))
b_f = np.array([0.1])  # Forget gate bias

W_i = np.array([0.7, 0.7])  # Input gate weight (updated shape to (2,))
b_i = np.array([0.2])  # Input gate bias

W_c = np.array([0.9, 0.9])  # Cell state candidate weight (updated shape to (2,))
b_c = np.array([0.05])  # Cell state candidate bias

W_o = np.array([0.6, 0.6])  # Output gate weight (updated shape to (2,))
b_o = np.array([0.15])  # Output gate bias

# Concatenated input (h_prev and x_t)
concat_input = np.concatenate((h_prev, x_t))

# Step 1: Forget Gate
f_t = sigmoid(np.dot(W_f, concat_input) + b_f)
print(f"Forget gate output f_t: {f_t}")

# Step 2: Input Gate
i_t = sigmoid(np.dot(W_i, concat_input) + b_i)
print(f"Input gate output i_t: {i_t}")

# Step 3: Cell State Candidate
C_t_candidate = tanh(np.dot(W_c, concat_input) + b_c)
print(f"Cell state candidate C_t_candidate: {C_t_candidate}")

# Step 4: Update Cell State
C_t = f_t * C_prev + i_t * C_t_candidate
print(f"Updated cell state C_t: {C_t}")

# Step 5: Output Gate
o_t = sigmoid(np.dot(W_o, concat_input) + b_o)
print(f"Output gate o_t: {o_t}")

# Step 6: Update Hidden State
h_t = o_t * tanh(C_t)
print(f"Updated hidden state h_t: {h_t}")
```
